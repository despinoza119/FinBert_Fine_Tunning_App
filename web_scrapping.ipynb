```python
import requests
from lxml import html
import pandas as pd

# URL of the Financial Times Markets page
url = 'https://www.ft.com/markets'

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    tree = html.fromstring(response.content)
    
    # Base XPath for the news items
    base_xpath = '//*[@id="stream"]/div[1]/ul/li'
    
    # Find the number of news items
    news_count = len(tree.xpath(base_xpath))
    
    # List to store news titles
    news_titles = []
    
    # Iterate over the news items and extract the text
    for i in range(1, news_count + 1):
        # Construct the XPath for the current item
        news_xpath = f'{base_xpath}[{i}]/div[2]/div/div/div[1]/div[2]/a'
        # Extract the text for the current news item
        news_item = tree.xpath(news_xpath)
        if news_item:
            news_text = news_item[0].text_content().strip()
            news_titles.append(news_text)
        else:
            print(f'News {i}: Could not retrieve the text.')
    
    # Create a DataFrame
    df = pd.DataFrame(news_titles, columns=["Financial Times News"])
else:
    print("Failed to retrieve the webpage")
```
